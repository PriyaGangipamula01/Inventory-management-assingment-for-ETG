```python
# READ THE DATA FROM JSON

import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
record
```

```python
#PURCHASING OF PRODUCTS

def purchase():
    ui_proid=input("Enter the product ID: ")
    ui_qn=int(input("Enter quantity: "))
    sales={}
    sales['pro_id']=ui_proid
    sales['name']=record[ui_proid]['name']
    sales['qn']=ui_qn
    sales['amount']=record[ui_proid]['pr']*ui_qn
    
    record[ui_proid]['qn'] = record[ui_proid]['qn'] - ui_qn
    
    return sales
    
cu_name=input("Enter your name: ")
cu_num=input("Enter your contact number: ")
num=0
sales={}
val=True
while val:
    print("1.You want to purchase an item")
    print("2.Finish shopping")
    choice=int(input("Enter your choice: "))
    if(choice==1):
        num=num+1
        sales[num]=purchase()
    elif(choice==2):
        val=False
    else:
        print("enter correct choice")
n=1
total=0
print("***************************")
print("Customer name:",cu_name)
print("Contact number:",cu_num)
print("***Your shopping details***")
print("You have purchased",num,"products")
for i in sales:
    print(n,"Product ID:",sales[i]['pro_id'])
    print("  Product name:",sales[i]['name'])
    print("  Quantity:",sales[i]['qn'])
    print("  Amount:",sales[i]['amount'])
    total=total+sales[i]['amount']
    n=n+1
print("Total amount:",total)
print("**************************")

    
js = json.dumps(sales)

fd = open("sales.json",'w')
fd.write(js)
fd.close()    
```

```python
#UPDATING INVENTORY AFTER PURCHASE

keys=record.keys()
for key in keys:
    if(record[key]['qn']==0):
        del record[key]
        break
        
js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()

record
```
